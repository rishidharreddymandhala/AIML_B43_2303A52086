```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
people = ['Arjun','Mani','Vikas']
age = [21,22,23]
height = [170,175,180]
weight = [70,75,80]
data = {'Name':people,'Age':age,'Height':height,'Weight':weight}
df = pd.DataFrame(data)
print(df)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Age'],df['Weight'],color = 'black',marker = '*')
plt.xlabel('Age (years)')
plt.ylabel('weight(kg)')
plt.title('Age vs Height')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(df['Name'],df['Height'],color = 'blue')
plt.title("Bar Chart of Individuals\' Height")
plt.xlabel("Name")
plt.ylabel("Height(cm)")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['Age'], bins=5, color='orange', edgecolor='black')
plt.title('Histogram of Age Distribution')
plt.xlabel('Age (years)')
plt.ylabel('Frequency')
plt.show()
```
